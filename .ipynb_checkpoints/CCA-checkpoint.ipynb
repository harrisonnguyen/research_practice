```python
def plot_result(X,Y):
    zero_class = np.where(Y[:, 0])
    one_class = np.where(Y[:, 1])
    plt.scatter(X[:, 0], X[:, 1], s=40, c='gray')
    plt.scatter(X[zero_class, 0], X[zero_class, 1], s=160, edgecolors='b',
                   facecolors='none', linewidths=2, label='Class 1')
    plt.scatter(X[one_class, 0], X[one_class, 1], s=80, edgecolors='orange',
                   facecolors='none', linewidths=2, label='Class 2')
```

```python
def plot_hyperplane(clf, min_x, max_x, linestyle, label):
    # get the separating hyperplane
    w = clf.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(min_x - 5, max_x + 5)  # make sure the line is long enough
    yy = a * xx - (clf.intercept_[0]) / w[1]
    plt.plot(xx, yy, linestyle, label=label)
```

```python
min_x = np.min(scikit_X[:, 0])
max_x = np.max(scikit_X[:, 0])

min_y = np.min(scikit_X[:, 1])
max_y = np.max(scikit_X[:, 1])
plot_hyperplane(classif.estimators_[0], min_x, max_x, 'k--',
                    'Boundary\nfor class 1')
plot_hyperplane(classif.estimators_[1], min_x, max_x, 'k-.',
                    'Boundary\nfor class 2')
plt.xticks(())
plt.yticks(())

plt.xlim(min_x - .5 * max_x, max_x + .5 * max_x)
plt.ylim(min_y - .5 * max_y, max_y + .5 * max_y)
plot_result(scikit_X,Y)
```

## perform my implementation of CCA

Source:
http://www.mitpressjournals.org/doi/pdf/10.1162/0899766042321814
1. We find the covariance matrix and split into $C_{xx}$, $C_{yy}$, $C_{xy}$ and $C_{yx}$
2. We solve the eigenproblem $ C_{xy}C_{yy}^{-1}C_{yx}\mathbf{w}_x = \lambda^2C_{xx}\mathbf{w}_x $ if $\mathbf{X}$ has the lower dimension.
3. We then solve $ \mathbf{w}_x = \frac{C_{xx}^{-1}C_{xy}\mathbf{w}_y}{\lambda}$ if $\mathbf{Y}$ is the lower dimension

```python
def closed_form_CCA(X,Y):
    z = np.concatenate((X.T,Y.T)) #combine the tranposed data. 
    #Each row represents a variable and column represents sample
    # X & Y must have the same number of samples
    C = np.cov(z) #find the covariance matrix of the variables (step 1)
    
    sx = X.shape[1] #find the dimensions of X and Y
    sy = Y.shape[1]
    
    #we partition the covariance matrix into the respective elements
    Cxx = C[0:sx,0:sx]
    Cxy = C[0:sx,sx:sx+sy]
    Cyx = Cxy.T
    Cyy = C[sx:,sx:]
    
    #we solve according to which dataset has the lower dimensions 
    if sy < sx:
        Z = np.dot(np.dot(np.dot(np.linalg.pinv(Cyy),Cyx),
                   np.linalg.pinv(Cxx)),Cxy)
        lam,wy = np.linalg.eig(Z) #solve the eigen problem for wy (step 2)
        lam = np.sqrt(np.real(lam))
        wx = np.dot(np.dot(np.linalg.pinv(Cxx),Cxy),wy)/lam #solve for wx (step 3)
    else:
        Z = np.dot(np.dot(np.dot(np.linalg.pinv(Cxx),Cxy),
                   np.linalg.pinv(Cyy)),Cyx)
        lam,wx = np.linalg.eig(Z) #solve the eigen problem for wx (step 2)
        lam = np.sqrt(np.real(lam))
        wy = np.dot(np.dot(np.linalg.pinv(Cyy),Cyx),wx)/lam #solve for wy (step 3)
    return lam,wx,wy    
```

```python
## A function that reduces the dimensionality of a dataset given 
## a dataset and a vector
def transform_data(data,vector):
    ## we normalise the data by subtracting the mean 
    ## and divide by standard deviation
    data -=np.mean(data,0)
    data /= np.std(data,0)
    
    #apply the transform
    data = np.dot(data,vector)
    return data
```

## Perform CCA using QR decomposition

http://numerical.recipes/whp/notes/CanonCorrBySVD.pdf

```python
def qr_CCA(X,Y,n_components = 2):
    #perform qr decomposition for X and Y
    q1,r1 = np.linalg.qr(X)
    q2,r2 = np.linalg.qr(Y)
    #perform SVD of q1 and q2
    U,s,Vh = np.linalg.svd(np.dot(q1.T,q2))
    
    a = np.dot(np.linalg.pinv(r1),U[:,:n_components])
    b = np.dot(np.linalg.pinv(r2),Vh.T[:,:n_components])
    return a,s[:n_components],b
```

# Perform kernel CCA on the data

```python
#create a kernel matrix for the data
def make_kernel(data,ktype = 'gaussian',gausigma = 1.0):
    if ktype == 'gaussian':
        from scipy.spatial.distance import pdist, squareform
        pairwise_dists = squareform(pdist(data, 'euclidean')) #compute pairwise distances
        kernel = np.exp(-pairwise_dists ** 2 / gausigma ** 2) #apply the RBF kernel
    elif ktype == 'linear':
        data =  data- np.mean(data,0)
        kernel = np.dot(data,data.T)
    return kernel
```

```python
def KCCA(X,Y,ktype = 'gaussian',epsilon = 1.0,n_components = 1):
    ndata_x, nfeature_x = X.shape
    ndata_y, nfeature_y = Y.shape
    
    #create the kernel matrix
    Kx = make_kernel(X,ktype=ktype)
    Ky = make_kernel(Y,ktype=ktype)
    
    I = epsilon * np.eye(ndata_x)
    
    #solve the eigemproblem
    KxI_inv = np.linalg.pinv(Kx + I)
    KyI_inv = np.linalg.pinv(Ky+I)
    
    L = np.dot(KxI_inv, np.dot(Ky, np.dot(KyI_inv,Kx)))
    U, s, Vh = scipy.linalg.svd(L)
    
    alpha = np.dot(KxI_inv, U[:, :n_components])
    beta = np.dot(KyI_inv, Vh.T[:, :n_components])
    
    return alpha,beta
```

```python
alpha,beta = KCCA(X,Y,ktype = 'gaussian',n_components = 2)
```

```python
classif = OneVsRestClassifier(SVC(kernel='linear'))
classif.fit(alpha, Y)
```

### Partial Gram-Schmidt Orthogonalisation

We use the PSGO algorithm to decompose our kernel matrix. This attempts to address the issue of using large training sets which may lead to computational problems and degeneracy. We use PSGO to apprxoimate the kernel matrices such that we are able to re-represent the correlation with reduced dimensionality

```python
def PSGO(data, nu):
    #apply the kernel
    kernel = make_kernel(data,'linear')
    #initialise the variables
    j = 0
    length = len(kernel)
    norm2 = np.zeros(length)
    index = np.zeros(length)
    size = np.zeros(length)
    feat = np.zeros((length,length))
    #find the norms of each data point
    for i in range(0,length):
        norm2[i] = kernel[i,i]

    while sum(norm2) > nu and j < length:
        ij = np.argmax(norm2) #get the index of the largest norm
        index[j] = ij
        size[j] = np.sqrt(norm2[ij]) #store the size
        
        for i in range(0,length):
            feat[i,j] = (kernel[i,ij] - np.dot(feat[i,0:j].T,feat[ij,0:j]))/size[j]
            norm2[i] = norm2[i] - feat[i,j]*feat[i,j]
        j +=1
    M = j
    return feat,M
```

## We try KCCA using incremental SVD

We follow the algorithm given in
http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.412.481&rep=rep1&type=pdf

"KERNEL CCA FOR MULTI-VIEW LEARNING OF ACOUSTIC FEATURES
USING ARTICULATORY MEASUREMENTS"

```python
def incremental_KCCA(X,Y,b = 200,r=100,t=1000):
    """
    b := block size, the number of columns of the kernel matrix that will be used
    r := rank-r of SVD
    t := number of columns of the kernel matrix that will be used
    """
    N = len(X)
    #initialise empty matrices
    Sx  = np.zeros((r,r))
    Ux = np.zeros((N,r))
    Vx = np.zeros((t+b,r))
    
    Sy  = np.zeros((r,r))
    Uy = np.zeros((N,r))
    Vy = np.zeros((t+b,r))
    
    for j in range(0,N,b):
        print "This is iteration number %d" %j
        #compute the jth to j+b columns of the kernel matrix
        Cx = np.zeros((b,N))
        #using the RBF kernel
        for i in range(0,N):
            Cx[:,i] = np.exp(-np.linalg.norm(
            (X[j:j+b,:] - X[i,:]),axis=1)**2/1.0**2)
        #perform the incremental SVD    
        Ux,Sx,Vx = incremental_SVD(Cx.T,Ux,Sx,Vx)
        
        Cy = np.zeros((b,N))
        #using the RBF kernel
        for i in range(0,N):
            Cy[:,i] = np.exp(-np.linalg.norm(
            (Y[j:j+b,:] - Y[i,:]),axis=1)**2/1.0**2)
        #perform the incremental SVD    
        Uy,Sy,Vy = incremental_SVD(Cy.T,Uy,Sy,Vy)
    
    F_hat = np.dot(np.sqrt(Sx), Ux.T)
    G_hat = np.dot(np.sqrt(Sy), Uy.T)
    
    Cff = np.dot(F_hat,F_hat.T)
    Cgg = np.dot(G_hat,G_hat.T)
    Cfg = np.dot(F_hat,G_hat.T)
    Cgf = np.dot(G_hat,F_hat.T)
    
    Cff_inv = np.linalg.inv(Cff)
    Cgg_inv = np.linalg.inv(Cgg)
    w, alpha_hat = np.linalg.eig(np.dot(np.dot(np.dot(Cff_inv,Cfg),Cgg_inv),Cgf))
    
    return np.dot(np.dot(np.linalg.inv(np.dot(F_hat.T,F_hat)),F_hat.T),alpha_hat)
```

```python
def incremental_SVD(C,U,S,V):
    t = 10 # temperorary
    b = C.shape[1]
    r = S.shape[0]
    
    L = np.dot(U.T,C) #compute projection
    #L is size (r x b)
    H = C-np.dot(U,L) #compute the residual
    #H is size (n x b)
    J,W = np.linalg.qr(H) #compute the QR decomposition
    #J is size (n x b), size W is (b x b)
    
    upper = np.concatenate((S,L),axis=1)
    lower = np.concatenate((np.zeros((b,r)),W),axis = 1)
    Q = np.concatenate((upper,lower),axis=0)
    #Q is size (r+b x r+b)
    
    #compute the SVD of Q
    U_approx,S_approx,Vt_approx = np.linalg.svd(Q)
    
    #perform the updates
    
    new_U = np.dot(np.concatenate((U,J),axis=1),U_approx)
    #new U is size (n x r+b)
    print V.shape
    upper = np.concatenate((V,np.zeros((t,b))),axis=1)
    lower = np.concatenate((np.zeros((b,r)),W),axis = 1)
    new_V = np.dot(np.concatenate((upper,lower),axis=0),Vt_approx.T)
    #new_V is size (t+b x r+b)
    #sort the singular values in S and corresponding singular vectors
    #in a descending
    sorted_index = np.argsort(S_approx)[::-1]
    
    new_U = new_U[:,sorted_index]
    new_V = new_V[:,sorted_index]
    new_S = np.diag(S_approx[sorted_index])
    
    #truncate the matrices and return the rank-r 
    return U[:,:r],S[:r,:r],V[:,:r]   
```
